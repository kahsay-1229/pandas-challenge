```python
import pandas as pd
#import os
```

```python
school_path = "Resources/schools_complete.csv"
student_path = "Resources/students_complete.csv"
```

```python
school_df = pd.read_csv(school_path)
school_df.head()
```

```python
student_df = pd.read_csv(student_path)
student_df.head()
```

```python
#calculate Total Schools using the uniq Sshool ID
Total_Schools = school_df['School ID'].count()
Total_Schools
#calculate Total Students using the uniq Student ID
Total_students = student_df['Student ID'].count()
Total_students
#calculate Total Budget(sum fun)
Total_Budget = school_df['budget'].sum()
Total_Budget
#calculate Avarage Math Score(mean fun)
Avarage_Math_Score = student_df['math_score'].mean()
Avarage_Math_Score
#calculate Reading Score(mean fun)
Avarage_Reading_Score =  student_df['reading_score'].mean()
Avarage_Reading_Score
#calculate % Passing Math
#Passing_Math = student_df
#calculate % Passing Reading
#calculate Overall Passing Rate (Average of the above two)
```

```python
print(Total_Schools)
print(Total_students)
print(Total_Budget)
print(Avarage_Math_Score)
print(Avarage_Reading_Score)
```

```python
#calculate % Passing Math
#find the math scores greater than 70
pass_math = student_df.loc[student_df['math_score']>=70]
pass_math.head()
```

```python
#calculate % Passing Math
math_percent = (len(pass_math)/(Total_students))*100
math_percent
```

```python
#calculate % Passing Reading
#find the reading scores greater that 70
pass_reading = student_df.loc[student_df['reading_score']>=70]
pass_reading.head()
```

```python
#calculate % Passing Math
read_percent = (len(pass_reading)/(Total_students))*100
read_percent
```

```python
#calculate Overall Passing Rate (Average of the above two)
overall_percent = (math_percent + read_percent)/2
overall_percent
```

```python
#create data frame for the District Summary
District_Summary =pd.DataFrame({"Total Schools": [Total_Schools],
                                "Total Students":[Total_students],
                                "Total Budget":[Total_Budget],
                                "Average Math Score": [Avarage_Math_Score],
                                "Average Reading Score": [Avarage_Reading_Score],
                                "% Passing Math": [math_percent],
                                "% Passing Reading": [read_percent],
                                "Overall Passing Rate": [overall_percent]

                               })
District_Summary
```

```python
# School Summary
# Set the school_df index to become the ‘school_name’ column and get type, total students, total budget and per student budget
school_index = school_df.set_index('school_name')
school_type = school_index['type']
school_students = school_index['size']
school_budget = school_index['budget']
psb = school_budget/school_students
school_index
```

```python
# Set school_name as index and groupby school_name for merge_df to get average scores by school
index = merge_data.set_index('school_name')
grouped = index.groupby(['school_name'])

# Get average math and reading score
school_avg_math = grouped['math_score'].mean()
school_avg_read = grouped['reading_score'].mean()
#print(school_avg_math)
#print(school_avg_read)
#grouped.head()
```

```python
# Calculate percent that passed math
total_stu = grouped['student_name'].count()
grouped_math = pass_math.groupby('school_name')
school_pass_math = grouped_math['student_name'].count()/total_stu*100
# Calculate percent that passed reading
grouped_reading = pass_reading.groupby('school_name')
school_pass_read = grouped_reading['student_name'].count()/total_stu*100

# Calculate overall passing rate
overall_pass = (school_pass_math + school_pass_read)/2
#print(total_stu)
#print(grouped_math)
#school_pass_math
```

```python
# Create dataframe for school summary
school_summary = pd.DataFrame({"School Type": school_type,
                              "Total Students": school_students,
                              "Total School Budget": school_budget,
                              "Per Student Budget": psb,
                              "Average Math Score": school_avg_math,
                              "Average Reading Score": school_avg_read,
                              "% Passing Math": school_pass_math,
                              "% Passing Reading": school_pass_read,
                              "Overall Passing Rate": overall_pass})
#school_summary 
```

```python
# Top Performing Schools (By Passing Rate) useing the loc method
top_summary = school_summary.loc[school_summary['Overall Passing Rate'] > 90]
top_summary.sort_values(['Overall Passing Rate'], ascending=False).head()
```

```python
# Bottom Performing Schools (By Passing Rate)
bottom_summary = school_summary.loc[school_summary['Overall Passing Rate'] < 80]
bottom_summary.sort_values(['Overall Passing Rate'], ascending=True).head()
```

```python
# check the type of all the columns in a DataFrame using the syntax dataFrameName[column_name].dtype or DF.dtypes attribute
school_summary["Per Student Budget"].dtype
school_summary.dtypes
```

```python
# create bins based on the per student budget spending 
budget_bins =[0, 500, 600, 700, 800]
# Create the names for the five bins
Spending_Ranges = ["<$500", "<$500-600", "<$600-700", "<$700-800"]
# using cut() function segregate 'Per Student Budget' elements into separate bins.
# cut( x, bins,right=True,labels=None,retbins=False,precision=3,include_lowest=False,duplicates="raise",)
# x:is the input array to be binned. It must be one-dimensional.
# bins: defines the bin edges for the segmentation.
# labels is used to specify the labels for the returned bins.                                  
school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"], budget_bins , labels=Spending_Ranges)
#spend_summary = school_summary.groupby("Spending Ranges (Per Student)")
#spend_summary = spend_summary[["Average Math Score", "Average Reading Score", "% Passing Math", 
                               #"% Passing Reading", "Overall Passing Rate"]]
#spend_summary.mean()
school_summary.head()
```

```python
spend_summary = school_summary.groupby("Spending Ranges (Per Student)")
spend_summary = spend_summary[["Average Math Score", "Average Reading Score", "% Passing Math", 
                               "% Passing Reading", "Overall Passing Rate"]]
spend_summary.mean()
```
