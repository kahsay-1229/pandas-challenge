```python
# Dependencies
import pandas as pd
#import os
```

```python
# Load in csv Files from resources 
school_path = "Resources/schools_complete.csv"
student_path = "Resources/students_complete.csv"
```

```python
# read and display CSV  files in pandas DataFrame. 
school_df = pd.read_csv(school_path)
student_df = pd.read_csv(student_path)
school_df.head()
```

```python
#  display student_df files in pandas DataFrame.
student_df.head()
```

```python
#  Check to see if there are any rows with missing data
student_df.count()
```

```python
# check data type. 
student_df.dtypes
```

```python
#calculate Total Schools using the uniq School ID
Total_Schools = school_df['School ID'].count()
Total_Schools
#calculate Total Students using the uniq Student ID
Total_students = student_df['Student ID'].count()
Total_students
#calculate Total Budget(use sum fun)
Total_Budget = school_df['budget'].sum()
Total_Budget
#calculate Avarage Math Score(use mean fun)
Avarage_Math_Score = round(student_df['math_score'].mean(),2)
Avarage_Math_Score
#calculate Reading Score(use mean fun)
Avarage_Reading_Score =  round(student_df['reading_score'].mean(),2)
Avarage_Reading_Score
```

```python
# print results to the consule
print(Total_Schools)
print(Total_students)
print(Total_Budget)
print(Avarage_Math_Score)
print(Avarage_Reading_Score)
```

```python
#calculate % Passing Math
#find the math scores greater than 70
pass_math = student_df.loc[student_df['math_score']>=70]
pass_math.head()
```

```python
#calculate % Passing Math
math_percent = round((len(pass_math)/(Total_students))*100, 2)
math_percent
```

```python
#calculate % Passing Reading
#find the reading scores greater that 70
pass_reading = student_df.loc[student_df['reading_score']>=70]
pass_reading.head()
```

```python
#calculate % Passing Math
read_percent = round((len(pass_reading)/(Total_students))*100, 2)
read_percent
```

```python
#calculate Overall Passing Rate (Average of the above two)
overall_percent = round((math_percent + read_percent)/2, 2)
overall_percent
```

```python
#create data frame for the District Summary using a dictionary
# use the map to cary the zero in the end  80.40, like (map, .2f)
District_Summary =pd.DataFrame({"Total Schools": [Total_Schools],
                                "Total Students":[Total_students],
                                "Total Budget":[Total_Budget],
                                "Average Math Score": [Avarage_Math_Score],
                                "Average Reading Score": [Avarage_Reading_Score],
                                "% Passing Math": [math_percent],
                                "% Passing Reading": [read_percent],
                                "Overall Passing Rate": [overall_percent]

                               })
District_Summary
```

```python
# to create a bine determine the min and max school_summary of per student budget
lowestPSB = school_summary['Per Student Budget'].min()
maxPSB = school_summary['Per Student Budget'].max()
print(lowestPSB)
print(maxPSB)
```

```python
# create bins based on the per student budget spending 
budget_bins =[500, 600, 700]
# Create the names for the five bins
Spending_Ranges = ["<$500-600", "<$600-700"]
# using cut() function segregate 'Per Student Budget' elements into separate bins.
# cut( x, bins,labels=None,)
# x:is the input array to be binned. It must be one-dimensional.
# bins: defines the bin edges for the segmentation.
# labels is used to specify the labels for the returned bins.                                  
school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"], budget_bins , labels=Spending_Ranges)
school_summary.head()
```

```python
spend_summary = school_summary.groupby("Spending Ranges (Per Student)")
spend_summary = spend_summary[["Average Math Score", "Average Reading Score", "% Passing Math", 
                               "% Passing Reading", "Overall Passing Rate"]]
spend_summary.mean()
```

```python
# to create a bine determine the min and max school_summary of Total Students
minstudent = school_summary['Total Students'].min()
maxstudent = school_summary['Total Students'].max()
print(minstudent)
print(maxstudent)
```

```python
# Scores by School Size
# create bins based on the Total Student column
studentscount_bin = [0,1000, 2000, 5000, ]
# create coulumn names for the studentscount_bin
schoolsize = ["Small", "Medium", "Large"]
#  using cut() function segregate 'Total Students' elements into separate bins.
school_summary["schoolsize"] = pd.cut(school_summary["Total Students"], studentscount_bin , labels=schoolsize)
school_capacity=school_summary.groupby("schoolsize")
school_capacity = school_capacity[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]
school_capacity.mean()
```
